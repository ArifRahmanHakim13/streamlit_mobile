```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
features = ['battery_power','clock_speed','mobile_wt','ram','int_memory','dual_sim','touch_screen','n_cores']
x = df[features]
y = df['price_range']
x.shape, y.shape
```

```python
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

```python
price_range_mapping = {
    0: 'Murah',
    1: 'Menengah Bawah',
    2: 'Menengah',
    3: 'Mahal'
}
```

```python
df['price_range'] = df['price_range'].map(price_range_mapping)
```

```python
df.to_csv('nama_file_hasil.csv', index=False)
```

```python
import numpy as np

# Data input
input_data = np.array([[1021, 0.5, 136, 2631, 53, 1, 1, 3]])

# Hasil prediksi dari model
prediction = lr.predict(input_data)

# Bulatkan hasil prediksi ke integer
rounded_prediction = int(round(prediction[0]))

# Membuat pemetaan nilai price_range
price_range_mapping = {
    0: 'Murah',
    1: 'Menengah Bawah',
    2: 'Menengah',
    3: 'Mahal'
}

# Memetakan hasil prediksi ke dalam deskripsi price_range
predicted_price_range = price_range_mapping[rounded_prediction]

# Output deskripsi price_range
print("Price Range: ", predicted_price_range)
```
